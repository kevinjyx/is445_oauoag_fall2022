```python
import matplotlib
import matplotlib.pyplot as plt
import datetime
import numpy as np
```

```python
time = [20070101, 20070401, 20070701, 20071001, 20080101, 20080401, 20080701, 20081001, 20090101]
```

```python
time
```

```python
plt.plot(time) #default matplolib line plot
plt.xlabel('Measurement #')
plt.ylabel('YearMonthDay')
```

```python
gdp = [14233.2, 14422.3, 14569.7, 14685.3, 14668.4, 14813.0, 14843.0, 14549.9, 14383.9]
gdp # gross domestic product (of USA) in Billions
```

```python
plt.plot(time, gdp, 'o-') # x-axis = time in YearMonthDay; y-axis=GDP in Billions
plt.xlabel('Time in YearMonthDay')
plt.ylabel('GDP in Billions')
```

```python
date = [] # empty array to fill with formatted dates
for t in time: 
    print(t)
    d = datetime.datetime.strptime(str(t),'%Y%m%d')
    print(d)
    date.append(d) # create a list of datetime-formatted objects
```

```python
d
```

```python
date
```

```python
plt.plot(date,gdp,'o-') # using datetime objects
```

```python
fig, ax = plt.subplots(figsize=(8,4)) # created fig (canvas) and axes objects (where we put specific plots)
ax.plot(date,gdp,'o-') # making a line+dot plot on a specific axis (could be multiple!)
plt.show() # showing our plot (for our purposes gets rid of the memory address stuff)
```

```python
style = 'fivethirtyeight'
with plt.style.context(style): # only apply this to this particular 
    fig, ax = plt.subplots(figsize=(8,4))
    ax.plot(date,gdp,'o-')
    plt.show()
```
