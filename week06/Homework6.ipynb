```python
import pandas as pd
import numpy as np
import ipywidgets
import matplotlib.pyplot as plt
from datetime import date
from bqplot import pyplot as plt
import bqplot
```

```python
licenses = pd.read_csv('https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_bcubcg_fall2022/main/data/licenses_fall2022.csv')
licenses = licenses.dropna(subset=['Effective Date','Expiration Date'])
```

```python
#licenses.head()
```

```python
#licenses.describe()
```

```python
licenses['Effective Date'] = pd.to_datetime(licenses['Effective Date'], errors='coerce')
licenses['Expiration Date'] = pd.to_datetime(licenses['Expiration Date'], errors='coerce')
licenses['Original Issue Date'] = pd.to_datetime(licenses['Expiration Date'], errors='coerce')
licenses['Original Issue Year'] = pd.DatetimeIndex(licenses['Original Issue Date']).month
licenses['Days Difference'] = (licenses['Expiration Date'] - licenses['Effective Date']).dt.days


```

```python
#licenses['Expiration Date']
```

```python
#licenses['Effective Date']
```

```python
#licenses['Days Difference']
```

```python
licenses['Means']=licenses['Days Difference']/2
```

```python
licenses['Means']
```

```python
chart = licenses.groupby(['License Status', 'License Type', 'Original Issue Year'], as_index=False)['Means'].sum()
chart
```

```python
table = pd.pivot_table(chart, values=['Means'], index=['License Type'],
                    columns=['License Status' ], aggfunc=np.sum)
```

```python
table
```

```python
mySelectedLabel = ipywidgets.Label()
```

```python
col_sc = bqplot.ColorScale(scheme="Reds") 
# this is because the "bins" are just bins -- their order is NOT numerically important
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()

# 3. Axis -- for colors, the axis is a colorbar!
ax_col = bqplot.ColorAxis(scale = col_sc, orientation='vertical', 
                          side='right') 
ax_x = bqplot.Axis(scale = x_sc, orientation='horizontal',tick_rotate = 270,
                   tick_style ={'font-size': 6}, label = 'License Status') # same x/y ax we had before
ax_y = bqplot.Axis(scale = y_sc, orientation='vertical', label='License Type')

anames = table.columns.levels[1].to_list()
for i in range(len(anames)):
    anames[i] = '      ' + anames[i]


# 4. Mark -- heatmap -- CHANGE HERE
heat_map = bqplot.GridHeatMap(color = np.log10(table.values),
                              row = table.index,
                              column = anames,
                              scales = {'color':col_sc, 
                                        'row':y_sc, 'column':x_sc},
                             interactions={'click':'select'},
                             anchor_style={'fill':'blue'},
                             selected_style={'opacity':1.0},
                             unselected_style={'opacity':0.9})
```

```python
# i,j = 2,16
i,j = 2,12

mask = (chart['License Type']== table.index[i]) & (chart['License Status']== table.columns.levels[1][j])
subset= chart[mask]
grouped = subset.groupby("Original Issue Year")["Means"].sum()

x_sch = bqplot.LinearScale() # range of z-axis data
y_sch = bqplot.LinearScale() # frequency of z-axis data in bins

# (2) axis: x/y
x_axh = bqplot.Axis(scale = x_sch, label = 'Original Issue Year')
y_axh = bqplot.Axis(scale = y_sch,  
                    orientation = 'vertical', 
                    label='Means')

line_plot = bqplot.pyplot.plot(x = grouped.index,
                               y = grouped.values, 
                              scales={'x':x_sch, 'y':y_sch})
```

```python
grouped
```

```python
def get_data_value(change):
    if len(change['owner'].selected) == 1: #only 1 selected
        i,j = change['owner'].selected[0]
        v = table.iloc[i,j]
        v2 = i,j
        mySelectedLabel.value = 'sum of total Means = ' + str(v) + f" {str(v2)}"# set our label
        mask = ((chart['License Type']== table.index[i]) & (chart['License Status']== table.columns.levels[1][j]))
        subset= chart[mask]
        grouped = subset.groupby("Original Issue Year")["Means"].sum()

        line_plot.x = grouped.index
        line_plot.y = grouped.values
heat_map.observe(get_data_value, 'selected')
```

```python
fig_heatmap = bqplot.Figure(marks = [heat_map], axes=[ax_col, ax_x, ax_y])
fig_line = bqplot.Figure(marks = [line_plot], axes = [x_axh, y_axh])
```

```python
fig_heatmap.layout.min_width='500px'
fig_line.layout.min_width='500px'

myDashboard = ipywidgets.VBox([mySelectedLabel, ipywidgets.HBox([fig_heatmap,fig_line])])
myDashboard
```
