import all necessary packages

```python
import pandas as pd
import numpy as np
import ipywidgets
import matplotlib.pyplot as plt
from datetime import date
from bqplot import pyplot as plt
import bqplot
```

import cvs from url and name it as licenses

```python
licenses = pd.read_csv('https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_bcubcg_fall2022/main/data/licenses_fall2022.csv')
licenses = licenses.dropna(subset=['Effective Date','Expiration Date'])
```

```python
licenses['Effective Date'] = pd.to_datetime(licenses['Effective Date'], errors='coerce')
licenses['Expiration Date'] = pd.to_datetime(licenses['Expiration Date'], errors='coerce')
licenses['Original Issue Date'] = pd.to_datetime(licenses['Expiration Date'], errors='coerce')
licenses['Original Issue Year'] = licenses['Original Issue Date'].dt.year
licenses['Days Difference'] = (licenses['Expiration Date'] - licenses['Effective Date']).dt.days
```

```python
licenses
```

form a table naming table

```python
table = np.log10(pd.pivot_table(licenses, values='Days Difference', index=['License Status'], columns=['License Type'], 
                      aggfunc = np.mean, fill_value=0))
table
```

```python
table.index.values.astype('str')
```

create the heat map using the above table data

### Heatmap

```python
# 2. Scales

col_sc = bqplot.ColorScale(scheme='Greens')
x_sc = bqplot.OrdinalScale(domain=table.index.values.astype('str').tolist())
y_sc = bqplot.OrdinalScale(domain=table.columns.values.astype('str').tolist())

# 3. Axis -- skip this for right now

col_ax = bqplot.ColorAxis(scale=col_sc,
orientation='vertical',
side='top')

ax_x = bqplot.Axis(label='License Status', scale=x_sc, tick_rotate=90) # tick_values=table.index.values.astype('str'))
ax_y = bqplot.Axis(label='License Type', orientation='vertical', scale=y_sc) # tick_values=table.columns.values.astype('str'),

# 4. Mark -- grid heatmap

heat_map = bqplot.GridHeatMap(color=np.log10(table.values), scales={'color':col_sc})

# put it all together as a figure

fig = bqplot.Figure(marks=[heat_map], axes=[col_ax, ax_x, ax_y])
fig
```

### barplot

```python
mask = (licenses['Original Issue Year'] < 2023) & (licenses['Days Difference'] > 0)
chart = licenses[mask].groupby(['Original Issue Year'], as_index=False)['Days Difference'].median()
chart
```

```python
# 2. scales
x_scl = bqplot.LinearScale()
y_scl = bqplot.LinearScale()

# 3. axis
x_axs = bqplot.Axis(label='Year', scale=x_scl)
y_axs = bqplot.Axis(label='Median Days Difference', scale=y_scl, orientation='vertical', side='left')

# 4. marks
hist_dur = bqplot.Bars(x=chart['Original Issue Year'], y=chart['Days Difference'],
                       scales={'x':x_scl, 'y':y_scl}, bins=chart.shape[0])

fig_dur = bqplot.Figure(marks=[hist_dur], axes=[x_axs, y_axs])
fig_dur
```

```python
myLabel = ipywidgets.Label()

def on_selected(change):
    if len(change['owner'].selected) == 1: # only selecting one bin at a time
        i,j = change['owner'].selected[0]
        chosen_status = table.index.tolist()[i]
        chosen_type = table.columns.tolist()[j]
        myLabel.value = 'chosen_type: {}, chosen_status: {}'.format(chosen_type, chosen_status)
        status_type_mask = (licenses['License Type']==chosen_type) & (licenses['License Status']==chosen_status)
        chart = licenses[mask & status_type_mask].groupby(['Original Issue Year'], as_index=False)['Days Difference'].median()
        # print(chart)
        hist_dur.x = chart['Original Issue Year']
        hist_dur.t = chart['Days Difference']
        

# 2. Scales

col_sc = bqplot.ColorScale(scheme='Greens')
x_sc = bqplot.OrdinalScale(domain=table.index.values.astype('str').tolist())
y_sc = bqplot.OrdinalScale(domain=table.columns.values.astype('str').tolist())

# 3. Axis -- skip this for right now

col_ax = bqplot.ColorAxis(scale=col_sc, orientation='vertical', 
                          side='top', tick_style={'font-size': 6})

ax_x = bqplot.Axis(label='License Status', scale=x_sc, tick_rotate=90, tick_style={'font-size': 6}) # tick_values=table.index.values.astype('str'))
ax_y = bqplot.Axis(label='License Type', orientation='vertical', scale=y_sc, tick_style={'font-size': 6}) # tick_values=table.columns.values.astype('str'),

# 4. marks
heat_map = bqplot.GridHeatMap(color=np.log10(table.values), 
                             scales={'color':col_sc},
                             interactions = {'click':'select'},
                             selected_style={'fill':'blue'})

# 5. interactivity
heat_map.observe(on_selected,'selected')

fig = bqplot.Figure(marks=[heat_map], axes=[col_ax, ax_x, ax_y])
#fig
#myDashboard = ipywidgets.VBox([myLabel,fig])
fig.layout.min_width='500px'
fig_dur.layout.min_width='500px'
figures = ipywidgets.HBox([fig,fig_dur])
myDashboard = ipywidgets.VBox([myLabel,figures])
myDashboard
```
