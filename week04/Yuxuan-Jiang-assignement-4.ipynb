Firstly, import the matplotlib.markers list from https://matplotlib.org/stable/api/markers_api.html

```python
list = [".", ",", "o", "v", "^", "<", ">", "1", "2","3", "4", "8", "s", "p", "P", "*", "h", "H", "+", "x", "X", "D", "d", "|", "_"]
```

Then, use the ipywidget package to set parameter and the interactive drop down menus. 

```python
@ipywidgets.interact(x_axis=bf.columns, y_axis=bf.columns, marker=list, color_bar=ipywidgets.ColorPicker())

def plot1(x_axis='date', y_axis='humidity', marker='.', color_bar=''):
    plt.plot(bf[x_axis], bf[y_axis],marker, color=color_bar)
    plt.title('The relationship')
    plt.xlabel(x_axis, fontsize=10)
    plt.ylabel(y_axis, fontsize=10)
    plt.show()
```

### Build a second widget that displays binned, aggregate values of a column where you can change:
### The field to "bin".  (Do you want the user to be able to pick any 2 columns?  Or only certain columns?  Why or why not?). You must support the selection of at least 1 categorical column and at least 1 numerical column.
### For numerical variables: The number of bins.  (Note: its OK if the binning widget is still present for non-numerical data, but it should have no effect on your plot).

```python
islider = ipywidgets.IntSlider(min=5,max=20,step=1, orientation='horizontal')
```

some code from week4's class note and website of https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.hist.html and https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html

In order to display binned, aggregate values of a column where I can change, I used the ipywidgets and slide bar. I used the bar chart since it may adjust the width and make each column presented independently. 

```python
@ipywidgets.interact(ColumnName=bf.columns, WidthNumber=islider)
def plot2(ColumnName='humidity', WidthNumber=5):
    plt.figure(figsize=(10,5))
    
    if bf[ColumnName].dtypes is object:
        index = bf[ColumnName].value_counts().index
        value = bf[ColumnName].value_counts().values
        plt.bar(index, value, width=0.5)
    else:
        plt.hist(bf[ColumnName], bins=WidthNumber)
    plt.xlabel(ColumnName, fontsize=10)
    plt.ylabel('Numbers', fontsize=10)
    plt.show()
```
